# 3rd pandas.read_csv with parse_dates, date_parser

```python
df = pd.read_csv("example.csv", index_col=["ID"], parse_dates=["TimeStamp"], date_parser=lambda t: pd.to_datetime(t, format="%Y-%m-%d %H:%M:%S"))
df
```

```python
df.dtypes
```

# 1st create num_friends

```python
users = [
    { "id": 0, "name": "Hero" },
    { "id": 1, "name": "Dunn" },
    { "id": 2, "name": "Sue" },
    { "id": 3, "name": "Chi" },
    { "id": 4, "name": "Thor" },
    { "id": 5, "name": "Clive" },
    { "id": 6, "name": "Hicks" },
    { "id": 7, "name": "Devin" },
    { "id": 8, "name": "Kate" },
    { "id": 9, "name": "Klein" }
]
friendships = [
    (0,1), (0,2),
    (1,2), (1,3),
    (2,3),
    (3,4),
    (4,5),
    (5,6), (5,7),
    (6,8),
    (7,8),
    (8,9)
]
```

```python
for user in users:
    user["friends"] =  []
users
```

```python
for i,j in friendships:
    users[i]["friends"].append(users[j]["id"])
    users[j]["friends"].append(users[i]["id"])
users
```

```python
for user in users:
    print(user["id"],user["friends"])
```

```python
def number_of_friends(user):
    return len(user["friends"])
```

```python
for user in users:
    print (user["id"], number_of_friends(user))
```

```python
num_friends = [number_of_friends(user) for user in users]
num_friends
```

# 3rd Count and Histogram of num_friends

```python
from collections import Counter
friend_counts = Counter(num_friends)
friend_counts
```

```python
# plot
xs = range(1,max(friend_counts)+1)
xs
```

```python
ys = [friend_counts[x] for x in xs]
ys
```

```python
max(xs) + 1
```

```python
from matplotlib import pyplot as plt
plt.bar(xs, ys)
plt.axis([0, max(xs)+1 , 0, max(ys)+1])
plt.xticks(xs)
plt.title("Histogram of user's Friends Counts")
plt.xlabel("# of friends")
plt.ylabel("# of people")
plt.rcParams["figure.figsize"] = (45,16)
plt.rcParams['axes.grid'] = True
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.show()
```

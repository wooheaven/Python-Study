```python
def get_max_pop_df(sido_name):
    max_pop = data.loc[data['광역시도'] == sido_name,'population'].max()
    return data.loc[data['population'] == max_pop,['광역시도','행정구역','population']]
```

```python
get_max_pop_df('경상북도')
```

```python
get_max_pop_df('서울특별시')
```

```python
max_pop_df = pd.DataFrame(columns=['광역시도','행정구역','population'])

for sido_name in data['광역시도'].unique():
    max_data = get_max_pop_df(sido_name)
    max_pop_df = max_pop_df.append(max_data)
    
max_pop_df = max_pop_df.reset_index()
max_pop_df.head()
```

```python
del max_pop_df['index']
max_pop_df.head()
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
%matplotlib inline
import seaborn as sns
sns.set()
rc('font', family='AppleGothic')
```

```python
figure, (ax1,ax2) = plt.subplots(nrows=2, ncols=1)
figure.set_size_inches(18,8)
sns.barplot(data=df_seoul, x="행정구역", y="population", ax=ax1)
sns.barplot(data=df_seoul, x="행정구역", y="면적", ax=ax2)
```

```python
df_seoul = df_seoul.sort_values(['population'],ascending=False).reset_index(drop=True)
df_seoul.head()
```

```python
figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,9)
sns.barplot(data=df_seoul, x="행정구역", y="population", ax=ax1)
ax1.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax1.set( ylabel='인구수')
for item in ax1.get_xticklabels(): 
    item.set_rotation(90)
for i, v in enumerate(df_seoul["population"].iteritems()):        
    ax1.text(i ,v[1], "{:,}".format(v[1]), color='m', va ='bottom', rotation=45)
plt.tight_layout()
```

```python
dicti=({'37':99943,'25':47228,'36':16933,'40':14996,'35':11791,'34':8030,'24' : 6319 ,'2'  :5055 ,'39' :4758 ,'38' :4611  })
pd_df = pd.DataFrame(list(dicti.items()))
pd_df.columns =["Dim","Count"]
pd_df.head()
```

```python
pd_df = pd_df.sort_values(['Count']).reset_index(drop=True)
pd_df.head()
```

```python
plt.figure(figsize=(12,8))
# plot barh chart with index as x values
ax = sns.barplot(pd_df.index, pd_df.Count)
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax.set(xlabel="Dim", ylabel='Count')
# add proper Dim values as x labels
ax.set_xticklabels(pd_df.Dim)
for item in ax.get_xticklabels(): item.set_rotation(90)
for i, v in enumerate(pd_df["Count"].iteritems()):        
    ax.text(i ,v[1], "{:,}".format(v[1]), color='m', va ='bottom', rotation=45)
plt.tight_layout()
plt.show()
```

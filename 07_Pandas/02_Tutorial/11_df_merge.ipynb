```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"key": list("bbacaab"), "data1": range(7)})
df1
```

```python
df2 = pd.DataFrame({"key": list("abd"), "data2": range(3)})
df2
```

# merge with column "key" like SQL inner join

```python
df_merge1 = pd.merge(df1, df2, on='key')
df_merge1
```

```python
df_merge2 = pd.merge(df2, df1, on='key')
df_merge2
```

# merge with column "key" like SQL full outer join

```python
df_merge3 = pd.merge(df1, df2, on='key', how='outer')
df_merge3
```

# merge with column "key" like SQL left outer join

```python
df_merge4 = pd.merge(df1, df2, on='key', how='left')
df_merge4
```

# merge with df3 column "key" and df4 index like SQL inner join

```python
df3 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 'value': range(6)})
df3
```

```python
df4 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
df4
```

```python
df_merge5 = pd.merge(df3, df4, left_on='key', right_index=True)
df_merge5
```

# merge with index like SQL full outer join

```python
df5 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]], index=['a', 'c', 'e'], columns=['Seoul', 'Incheon'])
df5
```

```python
df6 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13,14]], index=['b', 'c', 'd', 'e'], columns=['Daegu', 'Ulsan'])
df6
```

```python
df_merge6 = pd.merge(df5, df6, how="outer", left_index=True, right_index=True)
df_merge6
```

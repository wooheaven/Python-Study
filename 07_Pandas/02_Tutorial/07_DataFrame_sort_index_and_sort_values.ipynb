```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(6, 3))
df.columns = ["c", "b", "a"]
df.index = np.random.permutation(pd.date_range("20190111", periods=df.shape[0]))
df
```

```python
# df sort by row
df.sort_index()
```

```python
df.sort_index(axis=0, ascending=True) # default option
```

```python
df.sort_index(axis=0, ascending=False)
```

```python
# df sort by column
df.sort_index(axis=1)
```

```python
# df sort by row and column
df.sort_index().sort_index(axis=1)
```

```python
# df sort by column value
df.sort_values(by="b", ascending=False)
```

```python
# add e column
df["e"] = np.random.randint(0, 6, size=6)
df
```

```python
# add f column
df["f"] = ["A", "A", "B", "B", "C", "C"]
df
```

```python
# df sort by multiple column values
df.sort_values(by=["e", "f"])
```

```python
import pandas as pd
import numpy as np
```

```python
df_original = pd.read_csv('korea_gangwondo.csv')
df_original
```

```python
df1 = df_original[['shortName','population']]
df1 = df1.set_index('shortName')
df1
```

```python
df2 = df_original[['shortName', '면적']]
df2 = df2.set_index('shortName')
# df2 = df2.sort_values(by='면적', ascending=False)
df2
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
rc('font', family='AppleGothic')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)
fig.set_size_inches(18,8)
df1.plot(ax=axes[0], kind='bar', color='b')
df2.plot(ax=axes[1], kind='bar', color='r')
plt.tight_layout()
```

```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)
fig.set_size_inches(18,6)
# current_palette = sns.color_palette("hls", len(df1))
# sns.palplot(current_palette)
sns.barplot(data=df1, x=df1.index, y="population", ax=axes[0])
sns.barplot(data=df2, x=df2.index, y="면적", ax=axes[1])
axes[0].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
for i, v in enumerate(df1['population'].iteritems()):        
#     axes[0].text(i ,v[1], "{:,}".format(v[1]), color=current_palette[i], va ='bottom', rotation=45)
    axes[0].text(i ,v[1], "{:,}".format(v[1]), color='black', va ='bottom', rotation=45)
for i, v in enumerate(df2['면적'].iteritems()):
    axes[1].text(i ,v[1], "{:,}".format(v[1]), color='black', va ='bottom', rotation=45)
for item in axes[1].get_xticklabels(): 
    item.set_rotation(90)
plt.tight_layout()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(7, 7))
df.columns = ["a", "b", "c", "d", "e", "f", "g"]
df
```

```python
# replace NaN
df["c"] = [0, 0.827223, 1.249525, np.nan, 0.008320, np.nan, -0.828358]
df["d"] = [0, 0.827223, np.nan, 0.008320, np.nan, -0.828358, -0.494077]
df
```

```python
# remove row which has at least one NaN value
df.dropna(how="any")
```

```python
# remove row which has all values as NaN
df.dropna(how="all")
```

```python
# replace NaN
df.fillna(value=0.0)
```

```python
# remove rows which has NaN value by int index
df.loc[df.isnull()["c"], :]
```

```python
df.loc[df.isnull()["d"], :]
```

```python
target_index_list1 = df.loc[df.isnull()["c"], :].index.tolist()
target_index_list1
```

```python
target_index_list2 = df.loc[df.isnull()["d"], :].index.tolist()
target_index_list2
```

```python
target_index_list3 = target_index_list1 + [x for x in target_index_list2 if x not in target_index_list1]
target_index_list3.sort()
target_index_list3.reverse()
target_index_list3
```

```python
df
```

```python
df = df.drop(target_index_list3)
df
```

```python
df = df.set_index(np.arange(df.shape[0]))
df
```

```python
# remove rows which has NaN value by datetime index <- Don't need to reversed index list 
df = pd.DataFrame(np.random.randn(7, 7))
df.columns = ["a", "b", "c", "d", "e", "f", "g"]
df.index = pd.date_range("20190111", periods=df.shape[1])
df
```

```python
# replace NaN
df["c"] = [0, 0.827223, 1.249525, np.nan, 0.008320, np.nan, -0.828358]
df["d"] = [0, 0.827223, np.nan, 0.008320, np.nan, -0.828358, -0.494077]
df
```

```python
df.dropna(how="any")
```

```python
target_index_list4 = df.loc[df.isnull()["c"], :].index.tolist()
target_index_list4
```

```python
target_index_list5 = df.loc[df.isnull()["d"], :].index.tolist()
target_index_list5
```

```python
df = df.drop(target_index_list4 + target_index_list5)
df
```

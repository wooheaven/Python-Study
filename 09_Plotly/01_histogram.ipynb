```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"value1" : np.random.normal(0, 1, size=200), "value2" : np.random.normal(0,1, size=200)})
df.head()
```

```python
# plotly standard imports
import plotly.graph_objs as go
import plotly.plotly as py

# Cufflinks wrapper on plotly
import cufflinks

# Data science imports
import pandas as pd
import numpy as np

# Options for pandas
pd.options.display.max_columns = 30

# Display all cell outputs
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
from plotly.offline import iplot
cufflinks.go_offline()

# Set global theme
cufflinks.set_config_file(world_readable=True, theme='pearl')
```

```python
df['value1'].iplot(
    kind='hist',
    bins=30,
    xTitle='value',
    linecolor='black',
    yTitle='count',
    title='np.random.normal distribution')
```

```python
df['value1'].iplot(
    kind='hist',
    bins=30,
    xTitle='value',
    linecolor='black',
    yTitle='percentage (%)',
    histnorm='percent',
    title='np.random.normal distribution')
```

```python
df[['value1', 'value2']].iplot(
    kind='hist',
    linecolor='black',
    bins=20,
    histnorm='percent',
    bargap=0.1,
    opacity=0.8,
    barmode='group',
    xTitle='value',
    yTitle='percentage (%)',
    title='group of np.random.normal distributions')
```

```python
df[['value1', 'value2']].iplot(
    kind='hist',
    linecolor='black',
    bins=20,
    histnorm='percent',
    bargap=0.1,
    opacity=0.8,
    barmode='overlay',
    xTitle='value',
    yTitle='percentage (%)',
    title='overlay of np.random.normal distributions')
```

```python
df[['value1', 'value2']].iplot(
    kind='hist',
    linecolor='black',
    bins=20,
    histnorm='percent',
    bargap=0.1,
    opacity=0.8,
    barmode='stack',
    xTitle='value',
    yTitle='percentage (%)',
    title='overlay of np.random.normal distributions')
```
